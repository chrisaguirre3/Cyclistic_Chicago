```python
import pandas as pd
import numpy as np
import plotly.express as px
import pickle

pd.set_option('display.max_columns', None)
```

```python
# Define function to read the pickle that was written so that the object can be accessed upon kernel restart

def read_pickle(path, saved_object_name:str):
    '''
    In: 
        path:             path to folder where you want to read from
        saved_object_name: filename of pickled object you want to read in

    Out: 
        object: the pickled object
    '''
    with open(path + saved_object_name + '.pickle', 'rb') as to_read:
        object = pickle.load(to_read)

    return object
```

```python
# Read pickle

df1 = read_pickle(path, 'df1')
```

```python
df1_duplicated['start_end'] = np.where(df1_duplicated.index % 2, 'end', 'start')

df1_duplicated.head(10)
```

```python
df1_duplicated['avg_lat'] = np.where(df1_duplicated.index % 2, df1_duplicated['avg_end_lat'], df1_duplicated['avg_start_lat'])

df1_duplicated['avg_lng'] = np.where(df1_duplicated.index % 2, df1_duplicated['avg_end_lng'], df1_duplicated['avg_start_lng'])

df1_duplicated['station_name'] = np.where(df1_duplicated.index % 2, df1_duplicated['end_station_name'], df1_duplicated['start_station_name'])

df1_duplicated.head(10)
```
